```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv

np.random.seed(5)
```

```python
def addbias(dataframe):

    barr = [1.0] * dataframe.shape[0]
    # print(barr)
    dataframe["bias"] = barr

    # print(train_dataframe)
    return dataframe
```

```python
def predict(weights, x_test, y_test):
    # print(weights.shape)
    # print(x_test.shape)
    prediction = np.sign(np.dot(x_test, weights))
    return np.mean(prediction == y_test)*100
    # return prediction
```

```python
def avg_perceptron(x_train, y_train, iteration, learning_rate):

    num_features = len(x_train[0])
    # weights = np.zeros(num_features)
    weights = np.random.uniform(low=0.0, high=1.0, size = num_features)
    a = np.zeros(num_features)
    cnt = 0
    for _ in range(iteration):
        idx = np.random.permutation(len(x_train))
        x_train = x_train[idx]
        y_train = y_train[idx]
        for i in range(len(x_train)):
#             print(type(weights))
#             print(type(x_train[i]))
#             print(type(y_train))
            if y_train[i] * np.dot(x_train[i], weights) <= 0:
                weights = weights + learning_rate * y_train[i] * x_train[i]
            a = a + weights
            cnt+=1
    a = a/cnt
    return weights, a
```

```python
def k_fold_cross_validation():
    data_path = 'data/CVfolds/train{}.csv'
    learning_rate_arr = [1, 0.1, 0.01]
    print("=" * 100)

    print("Test and train set of different types ")
    std_acc = []
    dec_acc = []
    avg_acc = []
    for lr_rate in learning_rate_arr:

        std_lr_valu = 0
        dec_lr_valu = 0
        avg_lr_valu = 0
        for j in range(5):
            l = list(pd.read_csv(data_path.format(t)) for t in range(5) if t != j)
            train_dataframe = pd.concat(l).reset_index(drop=True)
            test_dataframe = pd.read_csv(data_path.format(j))
            # print(train_dataframe.shape, test_dataframe.shape)
            # summ.append(testing_with_depths(i, temp, temp_test))
            train_dataframe = addbias(train_dataframe)
            test_dataframe = addbias(test_dataframe)
            # print(train_dataframe.shape, test_dataframe.shape)
            train_x = train_dataframe.iloc[:, 1:].values
            train_y = train_dataframe.iloc[:, 0].values

            train_y[train_y == 0] = -1
            test_x = test_dataframe.iloc[:, 1:].values
            test_y = test_dataframe.iloc[:, 0].values
            test_y[test_y == 0] = -1


        # average perceptron
            lr_wts, a = avg_perceptron(train_x, train_y, 10, lr_rate)
            prediction = predict(a, test_x, test_y)
            avg_lr_valu = max(prediction, std_lr_valu)

        std_acc.append(std_lr_valu)
        dec_acc.append(dec_lr_valu)
        avg_acc.append(avg_lr_valu)

    # print(std_acc)
    std_max_acc = max(std_acc)
    idx = std_acc.index(std_max_acc)
    
    avg_max_acc = max(avg_acc)
    idx = avg_acc.index(avg_max_acc)

    print("Best hyper parameter for all cv folds for Average perceptron alg with accuracy as: {} is: {}".format(avg_max_acc,learning_rate_arr[idx]))


    return learning_rate_arr[idx]
```

```python
def eval_predict(weights, x_test):
   
    prediction = np.sign(np.dot(x_test, weights))
    return prediction
```

```python
def averaged_perceptron_call(train_x, train_y, test_x, test_y):
    print("*" * 100)
    print("Averaged Perceptron Algorithm")
    best_pred = 0
    best_lr = 0
    for learning_rate in [1, 0.1, 0.01]:
        learned_weights_averaged, averaged_weights = avg_perceptron(train_x, train_y, 10, learning_rate)
        prediction = predict(averaged_weights, train_x, train_y)
        # print("Train Accuracy of Averaged Perceptron with learning rate: {} is: {}".format(learning_rate, prediction))
        prediction = predict(averaged_weights, test_x, test_y)
        if(best_pred<prediction):
            best_pred = prediction
            best_lr = learning_rate
            
        print("Test Accuracy of Averaged Perceptron with learning rate: {} is: {}".format(learning_rate, prediction))

    return best_lr
```
